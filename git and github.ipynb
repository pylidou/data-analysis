Staging Area:
A place where we can group files together before we "commit" them to Git.

Commit
A "commit" is a snapshot of our repository. This way if we ever need to look back at the changes we've made (or if someone else does), we will see a nice timeline of all changes.


# check history

    git log

    git log --summary

# remote repo

    ssh-keygen -t rsa -C "youremail@example.com"

    git remote add origin https://github.com/yourname/yourname.git

# push remotely

    git push -u origin master

`origin`是远端名 ，而缺省是把本地的`master`发布到远端 。 ` -u`让Git记住这些参数，下次直接使用`git push`就可以

# pull remotely

假设现在其他人在github上添加了新文件

    git pull origin master


# difference

    git diff HEAD
    
#  staged difference

暂存区与工作区区别 

    git add octofamily/octodog.txt
    git diff --staged

# resetting the stage

You can unstage files by using the git reset command. Go ahead and remove octofamily/octodog.txt

    git reset octofamily/octodog.txt

# undo

    git checkout -- octocat.txt

    So you may be wondering, why do I have to use this '--' thing? git checkout seems to work fine without it. It's simply promising the command line that there are no more options after the '--'. This way if you happen to have a branch named octocat.txt, it will still revert the file, instead of switching to the branch of the same name.

# branch

    git branch clean_up

# switching branches

    git checkout clean_up

    git checkout -b clean_up

# remove all things

so you're in the clean_up branch. You can finally remove all those pesky octocats by using the git rm command which will not only remove the actual files from disk, but will also stage the removal of the files for us

    git rm '*.txt'

    git rm -r folder_of_cats

# commit branch changes

    git commit -m 'remove all the cats'

# switch back to master

    git checkout master

# merge

the moment has come when you have to merge your changes from the clean_up branch into the master branch. Take a deep breath, it's not that scary.

    git merge clean_up

# remove branch

    git branch -d clean_up

# git push

    git push
